```python
import os

def rename_images(folder_path, prefix="mkp_", start_index=1):
    # List all files in the specified folder
    files = os.listdir(folder_path)
    # Filter for common image file extensions
    image_files = [f for f in files if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif', '.tiff'))]

    # Loop through the image files and rename them
    for index, filename in enumerate(image_files, start=start_index):
        # Create the new name with zero-padded numbering
        new_name = f"{prefix}{index:03}{os.path.splitext(filename)[1]}"
        # Create full path for the old and new filenames
        old_path = os.path.join(folder_path, filename)
        new_path = os.path.join(folder_path, new_name)
        
        # Rename the file
        os.rename(old_path, new_path)
        print(f"Renamed '{filename}' to '{new_name}'")

# Specify the folder path where the images are located
folder_path = "images"
rename_images(folder_path)
```
